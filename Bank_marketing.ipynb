```python
from sklearn.preprocessing import StandardScaler,OrdinalEncoder # scaling numeric column,encoding categorical column
from sklearn.compose import make_column_selector # to select numeric and categoric columns
from sklearn.compose import ColumnTransformer #to transform columns using appropriate scaling / encoding
```

```python
target_column='y'
features=data.drop(columns=target_column)
target=data[target_column]
```

```python
from pandas.core.arrays import categorical
numeric_selector=make_column_selector(dtype_include='number')
categorical_selector=make_column_selector(dtype_exclude='number')
```

```python
numeric_columns=numeric_selector(features)
numeric_columns
```

```python
categorical_columns=categorical_selector(features)
categorical_columns
```

```python
data.head()
```

## train test split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=45,stratify=target)
```

```python
data['y'].value_counts(normalize=True)
```

```python
len(X_train)==len(y_train) # verify X and y have same length)
```

```python
len(X_test)==len(y_test)
```

```python
y_train.value_counts(normalize=True)
```

##preprocessing

```python
from sklearn.preprocessing import MinMaxScaler,OrdinalEncoder # scaling numeric column,encoding categorical column

preprocessor=ColumnTransformer([('category',OrdinalEncoder(),categorical_columns),
  ('numbers',MinMaxScaler(),numeric_columns)])

```

```python
preprocessor
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier

model=Pipeline([
    ('preprocess',preprocessor),
    ('dt_class',DecisionTreeClassifier())
])

model
```

## Hyperparameter Tuning

hyper parameters are parameters that are set before training .for example in decision tree max_depth,min_samples_split are hyperparameters. in practice,its hard to find optimal values for these parameters before training the model. the optimal values are found out by grid search cross-validation approach

```python
from sklearn.model_selection import GridSearchCV

param_grid ={'dt_class__max_depth':[2,4,3,5,7,9],
             'dt_class__min_impurity_decrease':[0.10,0.20,0.30]
             }# dt_class__ ---->to know preprocessor or dt_class ,it pass through the pipeline
```

```python
grid_search=GridSearchCV(model,param_grid,cv=5,refit=True)
```

```python
grid_search.fit(X_train,y_train)
```

```python
from sklearn.metrics import balanced_accuracy_score,precision_score,recall_score

best_model=grid_search.best_estimator_
y_train_pred=best_model.predict(X_train)
```

```python
balanced_accuracy=balanced_accuracy_score(y_train,y_train_pred)
precision=precision_score(y_train,y_train_pred,pos_label=1)
recall=recall_score(y_train,y_train_pred)
```
